**Import the necessary libraries**

```python
import pandas as pd
```

**Load and assign it to a variable called chipo.**

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'

hotel = pd.read_csv(url, sep = '\t')
```

```python
hotel.head()
```

**Sort the dataframe chipo by item_name column**

```python
hotel.sort_values('item_name')
```

**How many customers ordered Veggie Salad Bowl item?**

```python
hotel.item_name.value_counts()['Veggie Salad Bowl']
```

**How many times did someone order more than one Canned Soda?**

```python
len(hotel[(hotel.item_name =='Canned Soda') &  (hotel.quantity>1)])
```

**New Dataset**

```python
euro12 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv', sep=',')
euro12.head()
```

```python
euro12.shape
```

**Set Team columns as index in euro dataframe**

```python
euro12 = euro12.set_index('Team')
euro12
```

**Filter teams that scored more than 6 goals**

```python
euro12[euro12.Goals>6]
```

**Filter ['Denmark','Germany','Italy'] teams and ['Goals','Blocks'] columns from euro dataframe**

```python
euro12.loc[euro12.index.isin(['Denmark','Germany','Italy']), ['Goals','Blocks']]
```

**Present only the Shooting Accuracy from England, Italy and Russia**

```python
euro12.loc[euro12.index.isin(['England', 'Italy', 'Russia']), ['Shooting Accuracy']]
```
